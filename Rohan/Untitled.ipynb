```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d import axes3d
import pandas as pd

df = pd.read_csv("output.dat",sep='\t',header=None)
df.columns = ['a','b','sigma','root']

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df['a'], df['b'], df['root'], c='r', marker='o')

# Set the labels
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()

```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d import axes3d
import pandas as pd

df = pd.read_csv("output.dat",sep='\t',header=None)
df.columns = ['a','b','sigma','root']

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot a 3D sphere

# ax.plot_surface(df['a'], df['b'], df['root'], color='b')
ax.plot_trisurf(df['a'], df['b'], df['root'], linewidth=0, antialiased=False)


# Set the viewpoint
ax.view_init(elev=30, azim=60)

plt.show()
```
